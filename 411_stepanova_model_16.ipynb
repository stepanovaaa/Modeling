```python
import matplotlib.pyplot as plt
import random
import math

def v(x, y): 
  return math.sqrt(x * x + y * y)

m = 1000.0
g = 9.8
k = random.uniform(0, 1)

n = 0
h = 0.1

x = 0.0
y = 3000.0
x_1 = 1.0
y_1 = 1.0

x_array = [x]
y_array = [y]

while n > 10000 :
  x = x + h * x_1
  y = y + h * y_1
  vt = v(x_1, y_1)
  temp = -(k * vt) / m
  x_1 = x_1 + h * (temp * x_1)
  y_1 = x_1 + h * (temp * y_1 - g)
  if y > 0: 
    y_array.append(y)
    x_array.append(x)
  else: break
  n += h

plt.plot(x_array,
y_array)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

###2.1. Равномерно распределенная дискретная случайная величина

В сборочном цехе собирают изделия, состоящие из четырех элементов. Каждый элемент исправен с вероятностью 0,9. Изделие с исправными элементами посту-пает в продажу. Изделие с одним неисправным элементом ремонтируется. Из-делие с двумя или более неисправными элементами ремонту не подлежит. По-строить модель сборки изделий и определения числа неисправностей. Оценить вероятность того, что изделие работоспособно, изделие требует ремонта, изде-лие ремонту не подлежит. Оценку провести на основании 1000 испытаний.

```python
import random

sell = 0 # количество на продажу
repair = 0 # на ремонт
# изделия не подлежащие ремонту вычисляются 
# как разность всех изделий и на продажу и ремонт
n = 0 # испытание/изделие

while (n < 1000):
  k = 0 # количесво неисправных элементов в изделии
  # создаем изделие из 4-х элементов
  # неисправность элемента равна 1/10 - п.б.равно выпадению 0
  for i in range(4):
    element = random.randint(0, 9)
    if element == 0:
      k += 1
  # в зависимости от кол-ва неисправностей распределяем
  if k == 0:
    sell += 1
  if k == 1:
    repair += 1
  n += 1

# находим вероятность 
p_s = sell / n
p_r = repair / n
p_d = (n - sell - repair) / n

print(f"Оценка вероятности работоспособности = {p_s}")
print(f"Оценка вероятности ремонта = {p_r}")
print(f"Оценка вероятности утилизации = {p_d}")
```

```python
import random

T = random.uniform(8, 16)
t = T/8
print(f"T = {T}, t = {t}")
n = 0
k = 0

while n < 1000:
  # начальные точки интервала
  a1 = random.uniform(-t, T+t)
  a2 = random.uniform(-t, T+t)
  R1 = random.uniform(0, 1)
  R2 = random.uniform(0, 1)
  t1 = a1 + R1*T
  t2 = a2 + R2*T
  if abs(t2 - t1) <= t:
    k += 1
  n += 1

p = k / n
print(f"Оценка вероятности ожидания не больше t = {p}")
```

###2.3.Нормально распределенная случайная величина

Формула нормального распределения

```python
import random

arr = []
n = 0
while n < 3000:
  m = random.uniform(0.5, 1)
  q = random.uniform(0, m)
  R = 0
  k = 0
  while k < 12:
    R += random.uniform(0, 1)
    k += 1
  el = m + q * (R - 6)
  arr.append(el)
  n += 1

sum = 0
for i in range(n):
  sum += arr[i]

exp_value = sum / n
print(f'Математическое ожидание: {exp_value}')

sum = 0
for i in range(n):
  sum += (arr[i] - exp_value) ** 2

s_2 = sum / (n - 1)
print(f'Дисперсия: {s_2}')

s = s_2 ** (0.5)
print(f'Среднее квадратическое отклонение: {s}')
```

Система массового обслуживания состоит из одного обслуживающего
прибора и очереди неограниченной длины. Из источника требований в очередь
системы обслуживания поступает пуассоновский поток требований с интенсивностью λ . Постановка и выбор требований из очереди производятся прибором
в соответствии с дисциплиной PRIORITY (с приоритетом). Длительность обслуживания требований прибором является экспоненциально распределенной случайной величиной с параметром µ. Для выполнения условия существования стационарного режима полагаем λ < µ . Необходимо построить имитационную модель системы S. Модель должна вычислять оценки характеристик u и n системы обслуживания. 

Будем считать, что система S состоит из следующих объектов: источник
требований, очередь требований системы обслуживания, прибор системы обслуживания, требование. Все объекты системы S, за исключением требования,
могут быть только в единственном экземпляре. 

В нашем случае понадобятся следующие атрибуты требования модели: момент поступления требования tp в очередь системы из источника, момент начала обслуживания требования ts , момент завершения обслуживания требования tf, класс требования cl_req.

Заметим, что разность моментов tf и ts определяет длительность пребывания требования в системе обслуживания, а разность моментов ts и tp – длительность ожидания требования в очереди системы обслуживания. Для получения статистически значимых оценок характеристик u и b потребуется определенное число обслуженных требований. Организуем их хранение в модели следующим образом. 

Прибор, завершив обслуживание требования, будет направлять его не

Qобсл . Тогда признаком окончания эксперимента с имитационной моделью будет достижение заранее определенного числа обслуженных требований в очереди Qобсл . 

Установление требований в очереди и выбор их из очередей производится программными процессами, поэтому каждая очередь считается принадлежащей конкретному процессу. Основным набором характеристик очереди

и очередь Qобсл обслуженных требований. 

Модельное время в имитационной модели представлено глобальной переменной вещественного типа, принимающей значения на интервале [0, ∞) и обеспечивающей имитацию параллельного развития процессов системы S.
Обозначим Θ – текущий момент модельного времени.

Состояние модели системы S характеризуется совокупностью текущих

Событие – мгновенное изменение состояния модели системы S. Событие

* условиями возникновения;
* типом события, определяющим алгоритм обработки этого события;

В имитационной модели будем различать события трех типов: поступление требования в систему массового обслуживания, начало обслуживания требования прибором системы обслуживания и уход требования из системы массового обслуживания после завершения обслуживания. 

Процесс функционирования системы S в имитационной модели представим логически связанной последовательностью событий на оси модельного времени. Эта последовательность характеризуется интервалами времени между
